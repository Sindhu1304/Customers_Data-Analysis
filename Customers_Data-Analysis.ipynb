##**2.2 Removing Duplicate rows.**

##**2.3. Standardizing Phone number.**

```python
import re
```

```python
def standardize_phone(phone):

  digits = re.sub(r'\D', '', str(phone))
  if len(digits) == 10:
    return f"({digits[:3]}) {digits[3:6]}-{digits[6:]}"
  elif len(digits) == 11 and digits.startswith('1'):
        digits = digits[1:]
        return f"({digits[:3]}) {digits[3:6]}-{digits[6:]}"

  return phone

cleaned_data['Phone 1'] = cleaned_data['Phone 1'].apply(standardize_phone)
cleaned_data['Phone 2'] = cleaned_data['Phone 2'].apply(standardize_phone)

print("Standardized Phone Numbers:")
print(cleaned_data[['Phone 1', 'Phone 2']].head())
```

#**Task 3. Data Transformation.**

##**3.1. Creating new column as 'Full Name'.**

```python
cleaned_data["Full Name"] = cleaned_data["First Name"] + " " + ['Last Name']
cleaned_data.head()
```

##**3.2. Extracting email domain.**

```python
email_domains = cleaned_data['Email'].str.split('@').str[1]
print(email_domains)
```

```python
cleaned_data['Email Domain'] = email_domains
cleaned_data.head()
```

##**3.3. Converting date of birth column to datetime format.**

```python
cleaned_data['Subscription Date'] = pd.to_datetime(cleaned_data['Subscription Date'])
cleaned_data.head()
```

##**Calculating 'Live days'.**

```python
from datetime import datetime
cur_date = pd.to_datetime('today')
cleaned_data['Live Days'] = (cur_date - cleaned_data['Subscription Date']).dt.days
```

```python
cleaned_data.head()
```

#**Task 6 Data Visualization:**

##** 6.1 Create a bar chart showing the number of customers per country.**

```python
import matplotlib.pyplot as plt
x = cleaned_data['Country'].value_counts()
y = x.index

plt.figure(figsize=(45,10))
plt.bar(y,x)
plt.xlabel('Country')
plt.ylabel("Number of Customers")
plt.title("Number of Customers per Country")
plt.xticks(rotation=90, ha='right', fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

##**6.2 Generate a histogram displaying the distribution of customer ages.**

```python
import numpy as np
x = cleaned_data["Live Days"]

plt.figure(figsize=(12,8))
plt.hist(x, bins=15, color='purple')
plt.xlabel("Live Days")
plt.ylabel("Frequency")
plt.title("Distribution of Customer Ages")
plt.xticks(np.arange(1000,2000,500))

plt.show()
```

##** 6.3. Plot a pie chart representing the proportion of customers using different email domains.**


```python
x= cleaned_data["Email Domain"].value_counts()


plt.figure(figsize=(40,14))
plt.pie(x, labels=x.index, autopct='%0.1f%%')
plt.xlabel("Email Domains")
plt.ylabel("Percentage")
plt.title("Distribution of Email Domains")
plt.show()
plt.title("Distribution of Email Domains")
plt.show()
```
